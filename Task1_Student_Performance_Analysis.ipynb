## Step 1: Import Required Libraries
In this step, we import the necessary Python libraries.
**Pandas** is used for loading and analyzing the dataset.
**Matplotlib** is used for creating visualizations such as histograms, scatter plots, and bar charts.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 2: Load the Dataset
The student performance dataset is loaded into a Pandas DataFrame.
Since the dataset is separated by semicolons (`;`), we use the `sep=";"` parameter while reading the CSV file.

```python
df = pd.read_csv("student-mat.csv", sep=";")
```

## Step 8: Calculate Average Final Grade
The average final grade is calculated using the `mean()` function. This gives an overview of the overall academic performance of students.

```python
average_grade = df["G3"].mean()
print("Average final grade (G3):", average_grade)
```

## Step 9: Students Scoring Above 15
In this step, we identify students whose final grade (`G3`) is greater than 15 and count the total number of such students.


```python
df[df["G3"] > 15]
df[df["G3"] > 15].shape[0]
```

## Step 10: Study Time vs Final Grade
We calculate the Pearson correlation coefficient between study time (`studytime`) and the final grade (`G3`).
A correlation value close to +1 indicates a strong positive relationship, while a value close to 0 indicates little or no relationship.


```python
df["studytime"].corr(df["G3"])
```

## Step 11: Compare Performance by Gender
The dataset is grouped by gender (`sex`) to calculate the average final grade (`G3`) for male and female students. This helps compare their academic performance.

```python
df.groupby("sex")["G3"].mean()
```

## Step 12: Distribution of Final Grades
A histogram is created to visualize how students' final grades are distributed. It shows the frequency of each grade range.


```python
plt.hist(df["G3"])
plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.show()
```

## Step 13: Study Time vs Final Grade Scatter Plot
A scatter plot is used to visualize the relationship between study time and final grades. Each point represents one student.


```python
plt.scatter(df["studytime"], df["G3"])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Final Grade")
plt.show()
```

## Step 14: Average Grade by Gender
A bar chart is created to compare the average final grades of male and female students. This makes it easy to identify which group performs better on average.

```python
avg = df.groupby("sex")["G3"].mean()

avg.plot(kind="bar")
plt.title("Average Final Grade")
plt.show()
```
